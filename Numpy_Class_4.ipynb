```python
import numpy as np
```

```python
# Sum of two matrix.
x = np.array([[12,23,34],[45,56,67],[67,78,89]])
print(x)
print()
y = np.array([[34,45,56],[67,78,89],[54,43,33]])
print(y)
print()
z = x + y
print(z)
```

```python
# transpose
x = np.array([[12,23,34],[45,56,67]])
print(x)
print()
y = np.transpose(x)
print(y)
```

```python
# Matrix multiplication ----- for multiply use (.dot())
a = np.array([[12,23],[45,56]])
print(a)
print()
b = np.array([[34,45],[56,67]])
print(b)
print()
c = np.dot(a,b)
print(c)
```

```python
# Matrix Concatenation -- horizontal adding example 1 -- hstack always adding next to next.
x = np.array([[12,23],[45,67]])
print(x)
print()
y = np.array([[67,78],[65,54]])
print(y)
print()
print(np.hstack((x,y)))
```

```python
# example 2 --- horizontal 
x = np.array([[12],[34],[98]])
print(x)
print()
y = np.array([[34],[56],[78]])
print(y)
print()
print(np.hstack((x,y)))
```

```python
# Matrix Concatenation -- verticle adding example 1 -- vstack always adding down to next.
x = np.array([[12,23],[45,67]])
print(x)
print()
y = np.array([[67,78],[65,54]])
print(y)
print()
print(np.vstack((x,y)))
```

```python
# example 2 --- verticle
x = np.array([[12],[34],[98]])
print(x)
print()
y = np.array([[34],[56],[78]])
print(y)
print()
print(np.vstack((x,y)))
```

```python
# alternative ----- vstack 
x = np.array([[12],[34],[98]])
print(x)
print()
y = np.array([[34],[56],[78]])
print(y)
print()
print(np.concatenate((x,y),axis=0)) # axis = 0 means rows vstack.
```

```python
x = np.array([[12],[34],[98]])
print(x)
print()
y = np.array([[34],[56],[78]])
print(y)
print()
print(np.concatenate((x,y),axis=1)) # axis = 1 means columns hstack
```
