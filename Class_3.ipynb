```python
# Normal Copy -: In normal copy, if we want change in one copy then by default change another copy.
x = [2 ,4,8,4,5,8,6]
y = x
print(x)
print(y)
y[3] = "Abhisehk"
print(x)
print(y)


# Shallow Copy -: change in that copy only changes the value of that copy. it fails in nested list.
x = [2 ,4,8,4,5,8,6]
y = x.copy()
print(x)
print(y)
y[3] = "Abhishek"
print(x)
print(y)


x = [[12,23,34],[56,67,78]] # when it fails.
y = x.copy()
print(x)
print(y)
y[0][2]="Abhi"
print(x)
print(y)


# Deep Copy -: 
import copy
x = [[12,23,34],[56,67,78]]
y = copy.deepcopy(x)
print(x)
print(y)
y[0][2]="Abhi"
print(x)
print(y)

```

```python
# Filter -: used for extract all data.
# map -: map is used for changes in any data.
my_list = [1,5,4,6,8,11,3,12]
new_list = list(filter(lambda x: (x%2==0), my_list))
print(new_list)

li = [5,7,22,97,54,62,77,23,73,61]
final_list = list(filter(lambda x: (x%2!=0),li))
print(final_list)
```

```python
list2 = ["one", "TWO", "three", "FOUR"]
upper = list(filter(lambda x: x.isupper(), list2))
lower = list(filter(lambda x: x.islower(), list2))

print(upper)
print(lower)
print("-"*100)

list3 = ['one', 'two2', 'three3', '88', '99', '102']
numeric = list(filter(lambda x : x.isnumeric(), list3 ))
alpha = list(filter(lambda x : x.isalpha(), list3))
alphanum = list(filter(lambda x : x.isalnum(), list3))

print(numeric)
print(alpha)
print(alphanum)
```

```python
# Reduce -: using only two arguments and perform all program.
# when we want to use reduce function then first we need to import "from functools import reduce" 

doubles = [2,6,10,14,18]
from functools import reduce
def add(a,b):
    return a+b
sum_all = reduce(add,doubles) # reduce function will perform sum of all elements in the list.
sum_all
```

```python
import operator
```

```python
list2 = [1,2,3,4]
product = reduce (operator.mul, list2) # product of all numbers in a list2.
add = reduce (operator.add, list2) # add of all numbers in a list2.
concat_str = reduce (operator.add, ['Python', ' ', 'Rocks']) # concatenate string
prod = reduce (operator.mul, ['Hello', 3]) # Repeat a string multiple times.
min_num = reduce (lambda a, b: a if a < b else b, list2) # minimum number
max_num = reduce(lambda a, b: a if a > b else b, list2) # maximum number

print(product)
print(add)
print(concat_str)
print(prod)
print(min_num)
print(max_num)
```

```python
# Recursion -: a function call itself inside the definition.
def sum(n):
    if n == 0:
        return 0
    else:
        return (n+sum(n-1))

sum(5)



#   n = 5
#   n + sum(n-1)
#    5 + sum(4) --------------5 + 10 = 15
#     4 + sum(3) -------------4 + 6 = 10
#      3 + sum(2) ------------3 + 3 = 6
#       2 + sum(1) -----------2 + 1 = 3
#        1 + sum(0) ----------1 + 0 = 1
```

```python
# factorial 
def fact(n):
    if n == 0:
        return 1
    elif n ==1 :
        return 1
    else:
        return n*fact(n-1)

fact(6)

# n = 6
#  n*fact(n-1)
#   6*fact(5) ---------6*120
#    5*fact(4) ---------5*24
#     4*fact(3)  -------4*6
#      3*fact(2) -------3*2
#       2*fact(1) ------2*1
```
