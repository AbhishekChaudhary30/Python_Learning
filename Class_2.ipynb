```python
# Q1. Create a function that counts the number of vowels in a string
# Q2. Create a function that tells the sum of all the numbers between the minimum point to maximum point.

# Q1.
def my_count(a):
    c = 0
    for i in a:
        if i in "aeiouAEIOU":
            c+=1
    return c

my_count("Abhishek")
```

```python
# Q2. 
def my_sum(num):
    min_num = min(num)
    max_num = max(num)
    sum = 0
    for i in range(min_num, max_num + 1):
        sum+=i
    return sum

my_sum([1,2,3,4,5,6,7,8,9])    
```

```python
# Q2. 
def dis_card(a,b):
    s = ""
    for i in a:
        if i == b:
            s = s
        else:
            s = s+i
    return s

dis_card("Vishal","a")
```

```python
def dis_card_1(a,b):
    if isinstance(a,str):
        s = ""
        for i in a:
            if i == b:
                s = s
            else:
                s = s+i
        return s
    elif isinstance(a,list):
        l = []
        for i in a:
            if i != b:
                l.append(i)
        return l

l = [1,2,4,5,7,6]
dis_card_1(1,3)
dis_card_1("Abhishek","A")
```
