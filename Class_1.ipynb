```python
# create a password 
# it should be greater than 7
# It can not start with any number
# It contains a number
# third letter should be "T"
a = input("create a password")
if len(a) > 7 and a[0].isalpha() and a.isalnum() and a[2] == "T" and a[0].isupper():
    print("Password Accepted")
else:
    print("Please try again")
```

```python
# Check that the written name is valid or not 
# name should be bigger than 7 characters
# first letter should be in capital and 2nd should be in small
name = input("Enter your name here: ")
if len(name) > 7 and name[1].islower() and name[0].isupper(): 
    print("Name Accepted")
else:
    print("Not Accepted")
```

```python
rating = int(input("Please rate us: "))
review = input("Please Review us: ").lower()
if rating > 4:
    if review == "good" or review == "nice" or review == "best":
        print("Top rated")
    else:
        print("High rated")
else:
    if rating < 4:
        if review == "bad" or review == "poor" or review == "worst":
            print("Bottom rated")
    else:
        print("Low rated")
```
